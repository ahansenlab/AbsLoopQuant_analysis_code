```python
import numpy as np
import cooler
from matplotlib import pyplot as plt

# import the loop quantification module (contains main function to run AbLE)
import sys
sys.path.insert(1, '/mnt/md0/jjusuf/absloopquant/AbsLoopQuant_analysis_code')
import looptools
```

### Load files

```python
res = 1000  # bin size of Micro-C maps in bp

# open files for Npr3 rep1 Micro-C map (a.k.a. T1)

path_to_mcool = f'/mnt/coldstorage/jjusuf/20230912_MicroC/old/all_WTgenome/T1_WTgenome.mcool'  # available on GEO (GSE286495)
path_to_P_s_curve = '../AbsLoopQuant_data/P_s_T1_1000bp.txt'  # available in AbsLoopQuant_data folder; pre-calculated from calculate_p_s_curves.py (initial calculation can take several minutes)

clr = cooler.Cooler(path_to_mcool + f'::/resolutions/{res}')
P_s_values = np.loadtxt(path_to_P_s_curve)  # a numpy array containing the genome-wide averaged P(s) values at 0, 1, 2, 3, ... kb

# open files for mESC merged (15.6B) Micro-C map

path_to_mcool_mESC_merged = f'/mnt/coldstorage/jjusuf/Past_MicroC_Experiments/all_WTgenome/mESC_all_merged.mcool'  # available on GEO (GSE286495)
path_to_P_s_curve_mESC_merged = '../AbsLoopQuant_data/P_s_all_merged_1000bp.txt'  # available in AbsLoopQuant_data folder; pre-calculated from calculate_p_s_curves.py (initial calculation can take several minutes)

clr_mESC_merged = cooler.Cooler(path_to_mcool_mESC_merged + f'::/resolutions/{res}')
P_s_values_mESC_merged = np.loadtxt(path_to_P_s_curve_mESC_merged)  # a numpy array containing the genome-wide averaged P(s) values at 0, 1, 2, 3, ... kb
```

### Quantify loop

```python
# initialize loop quantifier object
lq = looptools.LoopQuantifier(clr, P_s_values)

# define loop position
chr_name = 'chr9'
start = 34_410_000
end = 34_663_000

# set key parameters
local_region_size = int(np.round(np.sqrt((end-start)/1000/(32/25**2)))) * 1000  # in bp; depends on loop size; see Methods
quant_region_size = 10000  # in bp; radius of circle used for quantification
k_min = 2  # parameter for outlier detection

# quantify the loop
AbLE_score = lq.quantify_loop(chr_name, start, end,
                              local_region_size=local_region_size,
                              quant_region_size=quant_region_size,
                              k_min=k_min,
                              clr_for_outlier_detection=clr_mESC_merged,
                              P_s_values_for_outlier_detection=P_s_values_mESC_merged)
```

### Plot loop

```python
# get the Micro-C matrix in the region of interest; loop +/- 20%
start_plot = int(start - (end-start)*0.2)
end_plot = int(end + (end-start)*0.2)
mat = clr.matrix().fetch(f'{chr_name}:{start_plot}-{end_plot}')
extent = np.array([start_plot, end_plot, end_plot, start_plot])/1e6  # for axis ticks, in Mb

# plot the matrix
fig, ax = plt.subplots()
plt.imshow(mat, cmap='Reds', extent=extent, vmin=0, vmax=0.0075)
plt.colorbar(label='interaction probability [arb.]');
plt.xlabel('position [Mb]')
plt.ylabel('position [Mb]')
plt.title(f'loop at {chr_name}:{start}-{end} (AbLE score = {AbLE_score:.4f})')
circ = plt.Circle((end/1e6, start/1e6), 0.010, facecolor='none', edgecolor='black', linewidth=1, linestyle='--')
ax.add_artist(circ);
```
